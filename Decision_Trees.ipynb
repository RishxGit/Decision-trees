```python
def compute_entropy(y):
    m = len(y)
    

    if m == 0:
        return 0.0
        
    p1 = np.sum(y) / m
    

    if p1 == 0 or p1 == 1:
        return 0.0
        
    entropy = -p1 * np.log2(p1) - (1 - p1) * np.log2(1 - p1)
    
    return entropy
```

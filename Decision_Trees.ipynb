```python
def compute_entropy(y):
    m = len(y)
    

    if m == 0:
        return 0.0
        
    p1 = np.sum(y) / m
    

    if p1 == 0 or p1 == 1:
        return 0.0
        
    entropy = -p1 * np.log2(p1) - (1 - p1) * np.log2(1 - p1)
    
    return entropy
```

```python
def split_dataset(X, node_indices, feature):
    left_indices = []
    right_indices = []
    for i in node_indices:
        if X[i, feature] == 1:
            left_indices.append(i)
        else:
            right_indices.append(i)
    return left_indices, right_indices
```
